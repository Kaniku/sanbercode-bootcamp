```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
import numpy as np
go_pivot=df.groupby('Genre')['NA_Sales','EU_Sales','JP_Sales','Other_Sales'].mean()
labeling=['NA_Sales','EU_Sales','JP_Sales','Other_Sales']
go_pivot
```

Expected Output :



<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
t=go_pivot.T
x=np.arange(len(labeling))
x1=np.arange(0,4,1)

y1=t['Action']
y2=t['Adventure']
y3=t['Fighting']
y4=t['Misc']
y5=t['Platform']
y6=t['Puzzle']
y7=t['Racing']
y8=t['Role-Playing']
y9=t['Shooter']
y10=t['Simulation']
y11=t['Sports']
y12=t['Strategy']

gambar,garis=plt.subplots(figsize=(30,19))
lebar=0.08
lebar1=0.08
garis.bar(x1,y1,lebar1,color='red',label='Action')
garis.bar(x1+lebar,y2,lebar1,color='blue',label='Adventure')
garis.bar(x1+lebar*2,y3,lebar1,color='purple',label='Fighting')
garis.bar(x1+lebar*3,y4,lebar1,color='gray',label='Misc')
garis.bar(x1+lebar*4,y5,lebar1,color='yellow',label='Platform')
garis.bar(x1+lebar*5,y6,lebar1,color='green',label='Puzzle')
garis.bar(x1+lebar*6,y7,lebar1,color='pink',label='Racing')
garis.bar(x1+lebar*7,y8,lebar1,color='red',label='Role-Playing')
garis.bar(x1+lebar*8,y9,lebar1,color='blue',label='Shooter')
garis.bar(x1+lebar*9,y10,lebar1,color='purple',label='Simulation')
garis.bar(x1+lebar*10,y11,lebar1,color='gray',label='Sports')
garis.bar(x1+lebar*11,y12,lebar1,color='yellow',label='Strategy')

plt.style.use('seaborn')
garis.set_xticklabels([0]+[' ']+['NA_Sales']+[' ']+['EU_Sales']+[' ']+['JP_Sales']+[' ']+['Other_Sales'])
plt.legend(bbox_to_anchor=[1.1,1.1],loc='best',prop={'size':25},facecolor='gray',frameon=True)
garis.set_title('Mean Sales Video Games By Genre',fontsize=30)
garis.set_xlabel('Region Sales',fontsize=20)
garis.set_ylabel('Mean Sales',fontsize=20)
plt.show
```

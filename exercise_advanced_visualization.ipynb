```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
gambar,garis = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True,figsize=(15,8))
garis[0].plot(x, tan, color= 'blue', label='tanx')
garis[1].plot(x, sin, color= 'red' ,label='sinx')
garis[2].plot(x, cos,color='green', label='cosx')
garis[0].set_title('Grafik tanx')
garis[1].set_title('Grafik sinx')
garis[2].set_title('Grafik cosx')
dapat_legenda=[garis.get_legend_handles_labels() for garis in gambar.axes]
dapat, legenda=[sum(lol,[])for lol in zip(*dapat_legenda)]
gambar.legend(dapat, legenda, loc='upper center')
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
from mpl_toolkits.axes_grid.inset_locator import inset_axes
gambar,garis= plt.subplots(figsize=(10,6))

garis.plot(x,y)
draw=inset_axes(garis,loc=2, width="30%", height=1.5, bbox_to_anchor=(-20,100,700,350))
draw.plot(x,y5, color='yellow')
draw.plot(x,y, color='blue')
draw1=inset_axes(garis,loc=1, width="30%", height=1.5, bbox_to_anchor=(100,100,700,350))
draw1.plot(x,y4, color='black')
draw1.plot(x,y, color='blue')
draw2=inset_axes(garis,loc=4, width="30%", height=1.5, bbox_to_anchor=(100,30,700,250))
draw2.plot(x,y3, color='green')
draw2.plot(x,y, color='blue')
draw3=inset_axes(garis,loc=3, width="30%", height=1.5, bbox_to_anchor=(-20,30,700,250))
draw3.plot(x,y2, color='red')
draw3.plot(x,y, color='blue')
garis.set_title('Grafik Cosx')
draw.set_title('Grafik CosX^5')
draw1.set_title('Grafik CosX^4')
draw2.set_title('Grafik CosX^3')
draw3.set_title('Grafik CosX^2')
plt.show()
```

<h2>2. Pandas Quiz: Menggabungkan DataFrame </h2>

Gabungkan Semua data dimana semua data nya berisi semua row dan column yang unik. berikan unique suffixes kepada nama column yang berlabel sama tapi berisi informasi yang berbeda.

Tidak boleh ada column yang berisi data informasi yang sama.

```python
# Code Here
supply_columns = df_category['Categories'].tolist()
all_columns = df_food.columns.tolist()
non_supply_columns = [col for col in all_columns if col not in supply_columns]
```

```python
df_non_supply_t = df_food[non_supply_columns]
df_food_t = df_food[['Country'] + supply_columns]
df_fat_t = df_fat[['Country'] + supply_columns]
df_protein_t = df_protein[['Country'] + supply_columns]
```

```python
all_df = pd.merge(df_food_t, df_fat_t, on='Country', how='outer', suffixes=('_food', '_fat'))
protein_new_cols = [col + '_protein' if col != 'Country' else col for col in df_protein_t.columns]
df_protein_t.columns = protein_new_cols
all_df = pd.merge(all_df, df_protein_t, how='outer', on='Country')
all_df = pd.merge(all_df, df_non_supply_t, how='outer')
```

<h2>4. Pandas Quiz: Binning Data, Grouping, dan Aggregating </h2>

```python
# Code Here
import numpy as np
goin=['Sangat Tidak Banyak','Tidak Banyak','Sedang','Cukup Banyak','Sangat Banyak']
df_food['Confirmed Category']=pd.qcut(df_food['Confirmed'],q=5,labels=goin)
group_by_confirmed=df_food.groupby('Confirmed Category').agg({'Obesity':np.mean,'Meat':np.mean,'Fruits - Excluding Wine':np.mean,'Fish, Seafood':np.mean,'Animal Products':np.mean,'Vegetal Products':np.mean})
group_by_confirmed
```

```python
group_by_confirmed.to_csv('groupby_confirmed.csv')
```

<h2>5. Matplotlib Visualization: Bar plot </h2>

Gunakan data grouping pada soal sebelumnya untuk melakukan visualisasi. jika kalian tidak dapat menyelesaikan soal tersebut, maka silahkan download data berikut yang merupakan hasil dari data grouping.

[Download](https://drive.google.com/uc?export=download&id=1ZHnO7O4yRUMiPmsX5zB_S_oB-w8lD1Yf)

Normalisasi data hasil grouping tersebut dengan rumus min-max normalisasi seperti di gambar. Kemudian buatlah grouping Visualization dengan berplot, visualisasikan data column 'Obesity', 'Fruits - Excluding Wine', 'Fish, Seafood', 'Animal Products' dan 'Vegetal Products'. 

Expected Output yang di hasilkan seperti gambar berikut.

![alt text](https://drive.google.com/uc?id=1v7aJMCeD_sUaZ0WQaQzAxCTy_EUJ7E4M)

```python
# Code Here
def cobafungsi(x):
    return(x-x.min())/(x.max()-x.min())
```

```python
knot=group_by_confirmed[['Obesity','Meat', 'Fruits - Excluding Wine', 'Fish, Seafood', 'Animal Products', 'Vegetal Products']].apply(cobafungsi)
knot
```

```python
import matplotlib.pyplot as plt
labeling=['Obesity','Meat', 'Fruits - Excluding Wine', 'Fish, Seafood', 'Animal Products', 'Vegetal Products']
t=knot.T
x=np.arange(len(labeling))
x1=np.arange(0,6,1)

y1=t['Sangat Tidak Banyak']
y2=t['Tidak Banyak']
y3=t['Sedang']
y4=t['Cukup Banyak']
y5=t['Sangat Banyak']

gambar,garis=plt.subplots(figsize=(30,19))
garis.set_facecolor('black')
lebar=0.08
garis.bar(x1,y1,lebar,color='blue',label='Sangat Tidak Banyak')
garis.bar(x1+lebar,y2,lebar,color='green',label='Tidak Banyak')
garis.bar(x1+lebar*2,y3,lebar,color='red',label='Sedang')
garis.bar(x1+lebar*3,y4,lebar,color='purple',label='Cukup Banyak')
garis.bar(x1+lebar*4,y5,lebar,color='orange',label='Sangat Banyak')

plt.style.use('seaborn')
garis.set_xticklabels([0]+['Obesity']+['Meat']+['Fruits - Excluding Wine']+['Fish, Seafood']+['Animal Products']+['Vegetal Products'])
legend=plt.legend(bbox_to_anchor=[0.9,0.95],loc='best',prop={'size':25},facecolor='black',frameon=True)
plt.setp(legend.get_texts(),color='white')
garis.set_title('Perbandingan Konsumsi Makanan Berdasarkan Tingkat Kasus COVID-19',fontsize=30)
plt.show
```

<h2>6. Matplotlib Visualization: Scatter Plot </h2>

```python
gbr, grs = plt.subplots(figsize=(15,6))
mapbar = grs.scatter(df_food['Vegetal Products'], df_food['Confirmed'], s=df_food['Deaths']*100000, alpha=0.6, cmap = 'gist_yarg',
                   c=df_food['Animal Products'])

gbr.colorbar(mapbar)
grs.set_facecolor('black')

grs.set_xlabel('Supply Vegetal Products')
grs.set_ylabel('Covid-19 case confirmed')
grs.set_title('Healthy Food and Covid-19')
plt.show()
```

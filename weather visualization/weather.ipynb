```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

```python
besar=df['WindAvgMPH']*20
warna=df['TempAvgF']
df.replace('-', np.nan, inplace=True)
df.fillna(method='ffill')
df['HumidityAvgPercent']=df['HumidityAvgPercent'].astype(float)
df['WindAvgMPH']=df['WindAvgMPH'].astype(float)
df['DewPointAvgF']=df['DewPointAvgF'].astype(float)

gambar,garis=plt.subplots(figsize=(15,6))

scatit=garis.scatter(df['HumidityAvgPercent'],df['DewPointAvgF'], alpha=0.6, c=warna,cmap='coolwarm',s=besar)
garis.set_xlabel('Humidity Avg %')
garis.set_ylabel('Dew Point Avg (F)')
garis.set_title('Austin Weather')
gambar.colorbar(scatit)
plt.show()
```
